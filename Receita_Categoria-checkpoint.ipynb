```python
#Título do projeto: Análise exploratória de dados de clientes.
#Objetivo: Demonstrar manipulação de dados com Pandas e geração de métricas básicas.
#Resultados:
#Idade média dos clientes.
#Faturamento total.
#Ticket médio por categoria.
#ferramentas utilizadas: Python, Pandas, Numpy.

#Observei também que para melhorar as vendas deve-se focar no marketing voltado para jovens na categoria livros 

import pandas as pd
import numpy as np
dados = pd.DataFrame({
    "cliente_id": [1, 2, 3, 4, 5, 6],
    "idade": [22, 35, 29, 41, 19, 33],
    "valor compra": [320, 1500, 230, 860, 120, 540],
    "categoria": ["Eletrônicos", "Eletrodomésticos", "Livros", "Eletrônicos", "Livros", "Eletrônicos"]
})
print(dados)
media_numpy = int(np.mean(dados["idade"]))
print("Média_idade (NumPy):", media_numpy)

total =  dados["valor compra"].sum()     
print("total compra",total)
medio_categoria =  dados.groupby("categoria")["valor compra"].mean().round(2) 
print(medio_categoria)
```
